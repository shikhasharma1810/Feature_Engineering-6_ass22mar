```python
Q1. Pearson correlation coefficient is a measure of the linear relationship between two variables. Suppose
you have collected data on the amount of time students spend studying for an exam and their final exam
scores. Calculate the Pearson correlation coefficient between these two variables and interpret the result.
```

```python
import pandas as pd

df = pd.DataFrame({'time of studying': [5.6,6,7.5,8,9,2.5],
              'final scores': [65,75,80,85,90,55]
             })
df
```

```python
df.corr(method='pearson')
```

```python
The value of Pearson correlation coefficient is in between -1 and +1.
The values of features time of studying and final scores are more towards +1, so,
the more positive corelated they both are.
```

```python
Q2. Spearman's rank correlation is a measure of the monotonic relationship between two variables.
Suppose you have collected data on the amount of sleep individuals get each night and their overall job
satisfaction level on a scale of 1 to 10. Calculate the Spearman's rank correlation between these two
variables and interpret the result.
```

```python
import pandas as pd

df = pd.DataFrame({'amount of sleep': [5.6,6,7.5,8,9,4],
              'job satisfaction level': [6,7,8,9,9,4]
             })
df
```

```python
df.corr(method='spearman')
```

```python
We observe from the spearman correlation being positive close to 1 symbolizes
the data is highly positvely correlated ie.with the number of sleep hours 
the overall job satisfaction level increases.
```

```python
Q3. Suppose you are conducting a study to examine the relationship between the number of hours of
exercise per week and body mass index (BMI) in a sample of adults. You collected data on both variables
for 50 participants. Calculate the Pearson correlation coefficient and the Spearman's rank correlation
between these two variables and compare the results.
```

```python
import numpy as np
import pandas as pd

hours_of_exercise = np.random.randint(1,20,size=50)
bmi = np.random.uniform(18,29,size=50)

df = pd.DataFrame({'Hours of exercise per week': hours_of_exercise ,
              'Body mass index': bmi})
df.head()
```

```python
# Pearson correlation coefficient

df.corr(method='pearson')
```

```python
#  Spearman's rank correlation

df.corr(method='spearman')
```

```python
We see there is a very little difference between the two correlation methods the 
values coming -0.096872 and -0.047709 for pearson and spearman respectively.

And we can see the values of the coefficient are close to -1, so, the features are
negatively corelated.
```

```python
Q5. A survey was conducted to examine the relationship between age and preference for a particular
brand of soft drink. The survey results are shown below:
    
Age(Years)     Soft drink Preference
  25              Coke
  42              Pepsi
  37              Mountain dew
  19              Coke
  31              Pepsi
  28              Coke
```

```python
import pandas as pd

df = pd.DataFrame({'Age(Years)': [25,42,37,19,31,28],
              'Soft drink Preference': ['Coke','Pepsi','Mountain dew','Coke','Pepsi','Coke']
             })
df
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder()
encoded = encoder.fit_transform(df[['Soft drink Preference']])
encoded_df = pd.DataFrame(encoded.toarray(),columns=encoder.get_feature_names_out())
df1 = pd.concat([df,encoded_df],axis=1)
df1
```

```python
df1.corr()
```
